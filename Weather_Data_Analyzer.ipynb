```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('weatherHistory.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['Temperature (C)'].fillna(df['Temperature (C)'].mean()) 
```

```python
df
```

```python
df = df[['Formatted Date','Temperature (C)','Humidity', 'Wind Speed (km/h)']]
```

```python
df
```

```python
# Convert once and handle timezone
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'], utc=True, errors='coerce')

# Extract features
df['Year'] = df['Formatted Date'].dt.year
df['Month'] = df['Formatted Date'].dt.month
df['Day'] = df['Formatted Date'].dt.day
df['Hour'] = df['Formatted Date'].dt.hour
```

```python
df.head()
```

```python
# Find hottest & coldest days
hotest = df.loc[df['Temperature (C)'].idxmax()]
coldest = df.loc[df['Temperature (C)'].idxmin()]
print("Hotest:  \n", hotest)
print("Coldest: \n", coldest)
```

```python
# Average Temp & Humidity
avg_temp = np.mean(df["Temperature (C)"])
avg_humidity = np.mean(df["Humidity"])

print("Average Temp:", avg_temp)
print("Average Humidity:", avg_humidity)
```

```python
# Line chart: Temperature trend
plt.figure(figsize= (10, 5))
plt.plot(df['Formatted Date'], df['Temperature (C)'], color= 'red')
plt.title('Daily Temperature Trend')
plt.xlabel('Year')
plt.ylabel('Temperature (°C)')
plt.show()
```

```python
# Histogram: Humidity distribution
plt.figure(figsize= (10, 5))
plt.hist(df["Humidity"], bins=20, color="skyblue", edgecolor="black")
plt.title("Humidity Distribution")
plt.xlabel("Humidity")
plt.ylabel("Frequency")
plt.show()
```

```python
# Bar chart: Average monthly temperature
plt.figure(figsize= (10, 5))
monthly_avg = df.groupby("Month")["Temperature (C)"].mean()
plt.bar(monthly_avg.index, monthly_avg.values, color="orange")
plt.title("Average Monthly Temperature")
plt.xlabel("Month")
plt.ylabel("Temperature (°C)")
plt.show()
```
